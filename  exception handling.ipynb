```python
import math
try:
    num = float(input("enter number:"))
    if num>0:
        print("ceil value:",math.ceil(num))
    elif num<0:
        print("floor value:",math.floor(num))
    else:
        print("number is zero")
except ValueError:
    print("invalid input ! please enter a valid number.")
```
